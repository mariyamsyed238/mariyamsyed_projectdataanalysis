## **Matplotlib Pyplot Module Import for Data Visuatization**

```python
 import matplotlib.pyplot as plt                # Importing the Plotly Express library for creating interactive plots and visualizations
```

```python
sentimentdataset=pd.DataFrame(df.head(20))
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.hist(sentimentdataset['Platform'], bins=10, color='blue')
plt.xlabel('Platform')
plt.ylabel('Frequency')
plt.title('Histogram of Platform')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(sentimentdataset['Platform'],sentimentdataset['Likes'],color='red')
plt.xlabel('Platform')
plt.ylabel('Likes')
plt.title('Platform over Likes')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(10,6))
plt.bar(sentimentdataset['User'], height=sentimentdataset['Hour'],color='green')
plt.xlabel('User')
plt.ylabel('Hour')
plt.xticks(rotation=45)
plt.title('Bar Plot of User')
```

```python
Platform_counts = sentimentdataset['Platform'].value_counts()
Platform_counts_sorted=Platform_counts.sort_index()
Platform_counts_sorted.plot(kind='line', color='pink')
plt.xlabel=('Platform')
plt.ylabel=('Frequency')
plt.xticks(rotation=45)
plt.title=('Frequency of Platform in dataset')
plt.tight_layout()
plt.show()
```

```python
Country_data=pd.DataFrame(df.tail(150))
plt.figure(figsize=(10,6))
Country_data['Country'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'pink', 'yellow', 'green'])

```

```python
data = {
    'User': ['User1', 'User2', 'User3', 'User4'],
    'Instagram': [100, 200, 150, 300],
    'Twitter': [50, 100, 80, 120],
    'Facebook': [80, 150, 100, 200]
    }
df = pd.DataFrame(data)
df.set_index('User', inplace=True)
plt.figure(figsize=(10, 6))
heatmap = plt.pcolor(df, cmap='YlGnBu')
plt.colorbar(heatmap)
plt.xticks(range(len(df.columns)), df.columns, rotation=45)
plt.yticks(range(len(df.index)), df.index)
plt.show()
```
