```python
df_cleaned = df_cleaned.drop_duplicates()         # Remove duplicate rows from the DataFrame
```

```python
df_cleaned.to_excel('cleaned_data.xlsx', index=False)
```

## **Matplotlib Pyplot Module Import for Data Visuatization**

```python
import matplotlib.pyplot as plt          # Importing the Plotly Express library for creating interactive plots and visualizations

```

```python
sample_superstore=pd.DataFrame(df.head(10))
```

## **Matplotlib Figure Setup with Custom Size**

```python
plt.figure(figsize=(10,6))
```

```python
plt.scatter(sample_superstore['City'],sample_superstore['Discount'], color='Green')
plt.xlabel('City')
plt.ylabel('Discount')
plt.title('Correlation between City and Discount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
```

```python
Category_counts = sample_superstore['Category'].value_counts()
```

```python
Category_counts_sorted=Category_counts.sort_index()
```

```python
Category_counts_sorted.plot(kind='line', color='black')
plt.xlabel=('Category')
plt.ylabel=('Frequency')
plt.xticks(rotation=45)
plt.title=('Frequency of Category in dataset')
plt.tight_layout()
plt.show()
```
