```python
import matplotlib.pyplot as plt
```

```python
sample_superstore=pd.DataFrame(df.head(10))
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.scatter(sample_superstore['City'],sample_superstore['Discount'], color='Green')
plt.xlabel('City')
plt.ylabel('Discount')
plt.title('Correlation between City and Discount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
```

```python
City_counts=sample_superstore['City'].value_counts()
```

```python
City_counts_sorted=City_counts.sort_index()
```

```python
City_counts_sorted.plot(kind='line', color='black')
plt.xlabel=('City')
plt.ylabel=('Frequency')
plt.xticks(rotation=45)
plt.title=('Frequency of city in dataset')
plt.tight_layout()
plt.show()
```
