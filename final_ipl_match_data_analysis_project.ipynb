```python
ipl_data = pd.DataFrame(df.head(10))

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(ipl_data['venue'], ipl_data['result_margin'], color='skyblue')
plt.xlabel('City')
plt.ylabel('Result Margin')
plt.title('Margin of Victory in IPL Matches Across Different venues')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
