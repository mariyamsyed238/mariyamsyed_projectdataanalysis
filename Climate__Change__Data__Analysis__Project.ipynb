## **Matplotlib Pyplot Module Import for Data Visuatization**

```python
import matplotlib.pyplot as plt              # Importing the Plotly Express library for creating interactive plots and visualizations
```

```python
df=pd.read_csv('/content/climate_change_indicators.csv')
```

```python
climate_data=pd.DataFrame(df.head(7))
```

## **Matplotlib Figure Setup with Custom Size**

```python
plt.figure(figsize=(10,6))
```

```python
plt.scatter(climate_data['Country'],climate_data['F1961'], color='blue')
plt.xlabel('Country'),
plt.ylabel('F1961')
plt.title('Correlation between Country and F1961')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
Country_counts=climate_data['Country'].value_counts()
```

```python
Country_counts_sorted=Country_counts.sort_index()
```

```python
plt.figure(figsize=(10,6))
```

```python
Country_counts_sorted.plot(kind='line',marker=0,color='black')
plt.xlabel('ISO3')
plt.ylabel('Frequency')
plt.title('Frequency of ISO3 in climate change')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```
