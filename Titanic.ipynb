```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/titanic.csv')
```

```python
# Handle missing values (e.g., drop or fill)
df = df.dropna()  # Drop rows with missing values
# or
df = df.fillna(method='ffill')  # Fill missing values using forward fill
```

```python
titanic=df.copy()                                       # Create a new DataFrame 'titanic' containing the same data as the existing DataFrame 'df'
```

```python
import matplotlib.pyplot as plt                # Importing the Plotly Express library for creating interactive plots and visualizations
```

```python
# Subset the DataFrame to include only the first 5 people
subset_titanic = titanic.head(5)
# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(subset_titanic['Name'], subset_titanic['SibSp'], color='red')
plt.xlabel('Name')
plt.ylabel('SibSp')
plt.title('Name over SibSp')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='Pclass')
plt.title('Number of Passengers by Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='Survived')
plt.title('Number of Survivors vs. Non-Survivors')
plt.show()
```

```python
survived_counts = titanic['Survived'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(survived_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Survivors vs. Non-Survivors')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic['Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='Pclass', y='Age')
plt.title('Age Distribution by Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=titanic, x='Survived', y='Age')
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
sns.pairplot(titanic.dropna(), hue='Survived', vars=['Age', 'Fare', 'Pclass'])
plt.show()
```

```python
# Filter numerical columns
numeric_df = titanic.select_dtypes(include=['float64', 'int64'])

# Compute and plot the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Titanic Features')
plt.show()
```
